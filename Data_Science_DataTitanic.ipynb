```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
df='/content/titanic.csv'
df=pd.read_csv(df)
df.head()
```

```python
df.shape
```

```python
# The number of people who survived: 342 survived
df['Survived'].sum()
```

```python
# The number of people who survived: 342
survived_filter = df['Survived']==1
df.loc[survived_filter, :]
```

```python
# The number of people who did not survived: 549 did not survive
didnot_survived_filter = df['Survived']==0
df.loc[didnot_survived_filter, :]
```

```python
# The percentage of people who survived: 38% (total = 891)
342/(342+549)
```

```python
survived_filter.mean()
```

```python
# How many of each sex survived?
# Total Female: 314 
# Total Male: 577
female_filter = df['Sex']=="female"
df.loc[female_filter, :]
```

```python
male_filter = df['Sex']=="male"
df.loc[male_filter, :]
```

```python
# How many of each sex survived?
# Survived and Female: 233 females of 314 females who survived. That is 233 females who survived out of 342 total people who survived. 
# Survived and Male: 109 males who survived
survivedAndFemale_filter = survived_filter & female_filter
df.loc[survivedAndFemale_filter, :]
```

```python
# Number of only males who survived: 109
survivedAndMale_filter = survived_filter & male_filter
df.loc[survivedAndMale_filter, :]
```

```python

df['Fare']<10
```

```python
# The group of passengers who paid less than $10, what percentage of this group survived: 336
fare_lessThan10=df['Fare']<10
df.loc[fare_lessThan10, :]
```

```python
# Overall, what is the average age of those who didn't survive? 31 the average age of those who did not survived
didnot_survivedAndAge_filter= didnot_survived_filter & df['Age']
df.loc[didnot_survivedAndAge_filter, :]
```

```python
# Overall, what is the average age of those who didn't survive? 31 the average age of those who did not survived
df.loc[didnot_survivedAndAge_filter, 'Age'].mean()
```

```python
# Survived and Age:
survivedAndAge_filter= survived_filter & df['Age']
df.loc[survivedAndAge_filter, :]
```

```python
# Overall, what is the average age of those who did survive? 28 year old who survived
df.loc[survivedAndAge_filter, 'Age'].mean()
```

```python
# What is the average age of those who did and didn't survive grouped by Sex?
df.groupby(['Survived', 'Sex'])[['Age']].mean()
```
