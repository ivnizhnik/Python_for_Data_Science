```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
bike_df=pd.read_excel('/content/RailsToTrails_National Count Data_week.xlsx')
bike_df.head()
```

```python
# Select just data where the 'Change 2019-2020' column is greater than 100% (greater than 1).
change_filter = bike_df['Change 2019-2020'] > 1
```

```python
bike_df.loc[change_filter, :].head()
```

```python
# Select just the data after July 31, 2020.
time_filter = bike_df['Week of'] > '2020-07-31'
```

```python
bike_df.loc[time_filter, :].head(10)
```

```python
# Put these together! Where there any weeks in the second half of the year (after July 31, 2020) where the change was greater than 100%?
timechange_filter = time_filter & change_filter
```

```python
bike_df.loc[timechange_filter, :].head()
```

```python
# No. There are NO any weeks in the second half of the year (after July 31, 2020) where the change was greater than 100%
```
