```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (12, 10))
time = [ 2017 , 2013 , 2008 , 2007 , 2006 , 2005, 2004 ]  # x-axis
football = [ 37 , 39, 41, 43, 43 , 34	 , 37 ]   # x-axis
basketball =	[11, 	12,	9,	11,	12,	12,	13]   # x-axis
baseball= [9,	14,	10,	13,	11,	12,	10]
soccer =[7,	4,	3,	2,	2,	3,	2]	
df_time_sports = pd.DataFrame(
    { 'time': time,
     'football': football , 
     'basketball': basketball,
     'baseball': baseball,
     'soccer' : soccer,})
ax = plt.gca()
plt.title('What sports are most watched in the US over time?', fontsize = 25)
df_time_sports.plot( x = 'time' , y = 'football', fontsize = 15, ax = ax )
df_time_sports.plot( x = 'time' , y = 'basketball', ax = ax )
df_time_sports.plot( x = 'time' , y = 'baseball', ax = ax )
df_time_sports.plot( x = 'time' , y = 'soccer', ax = ax );
```

```python
# According to the graph above, we see an increase in soccer and a decrease in football over time. With that noticed, the football is still the most watched sport in the US.
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (15, 12))
time = [ 2017 , 2013 , 2008 , 2007 , 2006 , 2005, 2004 ]  # x-axis
football = [ 37 , 39, 41, 43, 43 , 34	 , 37 ]   # x-axis
basketball =	[11, 	12,	9,	11,	12,	12,	13]   # x-axis
baseball= [9,	14,	10,	13,	11,	12,	10]
soccer =[7,	4,	3,	2,	2,	3,	2]	
ice_hockey = [4,	3, 4,	4,	2,	4,	3]
auto_racing	= [2,	2,	3,	3,	4,	5,	5]
tennis =[2,	3,	1,	1,	1,	3,	2]
golf	=[1,	2,	2,	2,	3,	2,	2]
other	=[5,	4,	6,	3,	3,	4,	4]
none	= [15, 11, 14,	12,	12,	13,	12 ]
df_time_sports = pd.DataFrame(
    { 'time' : time , 
     'football': football , 
     'basketball': basketball,
     'baseball': baseball,
     'soccer' : soccer,
     'ice_hockey' : ice_hockey,
     'auto_racing': auto_racing,
     'tennis': tennis,
     'golf': golf,
     'other' : other,
     'none' : none})
ax = plt.gca()
plt.title('What sports are most watched in the US over time?', fontsize = 25)
df_time_sports.plot( x = 'time' , y = 'football', ax = ax )
df_time_sports.plot( x = 'time' , y = 'basketball', ax = ax )
df_time_sports.plot( x = 'time' , y = 'baseball', ax = ax )
df_time_sports.plot( x = 'time' , y = 'soccer', ax = ax )
df_time_sports.plot( x = 'time' , y = 'ice_hockey', ax = ax )
df_time_sports.plot( x = 'time' , y = 'auto_racing', ax = ax )
df_time_sports.plot( x = 'time' , y = 'tennis', ax = ax )
df_time_sports.plot( x = 'time' , y = 'golf', ax = ax )
df_time_sports.plot( x = 'time' , y = 'other', ax = ax )
df_time_sports.plot( x = 'time' , y = 'none', ax = ax );
```
